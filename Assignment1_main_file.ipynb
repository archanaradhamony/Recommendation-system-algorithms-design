```python
import load_dataset_mod
import similarity_module


def main():
    print ('\n**************************************************************************************************')
    print('******************* WELCOME TO THE ARTIST AND MUSIC RECOMMENDATION SYSTEM*************************')
    print ('**************************************************************************************************\n')
    #Music features dictionary is a nested dictionary of music name and corresponding attributes
    #artist_music dictionary has artist name as key and all the music names by the artist, as a list assigned to each artist.
    music_features = {}
    music_features = load_dataset_mod.build_music_features()  
    artist_music = {}
    artist_music = load_dataset_mod.build_artist_music()
    artist_music_feature = {}
    try:
        choice = int(input('Press 1 if you like to compare between 2 artists OR Press 2 if you like to compare between 2 music'))
        if choice == 1:
            artist1 = input('Please enter the name of the 1st artist: ')
            artist2 = input('Please enter the name of the 2nd artist: ')
            #comparison between artists. 
            #artist_music_feature dictionary is created, to find the average value of each attribute, for each artist, for all the music by the artist.
            artist_music_feature = load_dataset_mod.build_artist_music_features(artist1, artist2, artist_music, music_features)
            print(artist_music_feature)
            eucilidean_choice = int(input('\nWould you like to know the Eucilidean Distance? Please press 1 if so\n'))
            if (eucilidean_choice == 1):
  #              print('in eucilidean')
                print('Eucilidean Distance is: ', similarity_module.eucilidean_distance(artist1, artist2, artist_music_feature))
            else: 
                Continue
            
            cosine_choice = int(input('\nWould you like to know the Cosine Distance?Please press 1 if so\n'))
            if (cosine_choice == 1): 
                print('cosine Distance is: ', similarity_module.cosine_distance(artist1, artist2, artist_music_feature))
                if float(similarity_module.cosine_distance(artist1, artist2, artist_music_feature)) > 0.9:
                    print('As per Cosine Similarity, there is a good match')
            else: 
                Continue
            
            jaccard_choice = int(input('\nWould you like to know the Jaccard Distance? Please press 1 if so\n'))
            if (jaccard_choice == 1):
                print('Jaccard Distance is: ', similarity_module.jaccard_distance(artist1, artist2, artist_music_feature))
            else: 
                Continue
            
            manhattan_choice = int(input('\nWould you like to know the Manhattan  Distance? Please press 1 if so\n'))
            if (manhattan_choice == 1):
                print('Manhattan Distance is: ', similarity_module.manhattan_distance(artist1, artist2, artist_music_feature))
            else: 
                Continue            
                
            pearson_choice = int(input('\nWould you like to know the Pearson  Distance? Please press 1 if so\n'))
            if (pearson_choice == 1):
                print('Pearson Distance is: ', similarity_module.pearson_coefficient(artist1, artist2, artist_music_feature))
                if float(similarity_module.pearson_coefficient(artist1, artist2, artist_music_feature)) > 0.9:
                    print('As per Pearson Similarity, there is a good match')
                elif float(similarity_module.pearson_coefficient(artist1, artist2, artist_music_feature)) == 0:
                    print('As per Pearson Similarity, there is no relationship')
                elif float(similarity_module.pearson_coefficient(artist1, artist2, artist_music_feature)) <=(-1):
                    print('There is a negative correlation')
            else: 
                Continue
        elif choice ==2:
            music1 = input('Please enter the name of the 1st music: ')
            music2 = input('Please enter the name of the 2nd music: ')
            
            eucilidean_choice = int(input('\nWould you like to know the Eucilidean Distance? Please press 1 if so\n'))
            if (eucilidean_choice == 1):
                print('in eucilidean')
                print('Eucilidean Distance is: ', similarity_module.eucilidean_distance(music1, music2, music_features))
            else: 
                Continue
            
            cosine_choice = int(input('\nWould you like to know the Cosine Distance?Please press 1 if so\n'))
            if (cosine_choice == 1): 
                print('cosine Distance is: ', similarity_module.cosine_distance(music1, music2, music_features))
                if float(similarity_module.cosine_distance(music1, music2, music_features)) > 0.9:
                         print('As per Cosine Similarity, there is a good match')
            else: 
                Continue
            
            jaccard_choice = int(input('\nWould you like to know the Jaccard Distance? Please press 1 if so\n'))
            if (jaccard_choice == 1):
                print('Jaccard Distance is: ', similarity_module.jaccard_distance(music1, music2, music_features))
            else: 
                Continue
            
            manhattan_choice = int(input('\nWould you like to know the Manhattan  Distance? Please press 1 if so\n'))
            if (manhattan_choice == 1):
                print('Manhattan Distance is: ', similarity_module.manhattan_distance(music1, music2, music_features))
            else: 
                Continue            
                
            pearson_choice = int(input('\nWould you like to know the Pearson  Distance? Please press 1 if so\n'))
            if (pearson_choice == 1):
                print('Pearson Distance is: ', similarity_module.pearson_coefficient(music1, music2, music_features))
                if float(similarity_module.pearson_coefficient(music1, music2, music_features)) > 0.9:
                    print('As per Pearson Similarity, there is a good match')
                elif float(similarity_module.pearson_coefficient(music1, music2, music_features)) ==0:
                    print('As per Pearson Similarity, there is no relationship')
                elif float(similarity_module.pearson_coefficient(music1, music2, music_features)) <=(-1):
                    print('There is a negative correlation')
                
            else: 
                Continue
    except ValueError:
        print('wrong input')
    except:
        print('unknown error')
    else:
        print('All good')


main()

#artist_music_feature = {}
#artist_music_feature = load_dataset_mod.build_artist_music_features('Carl Woitschach', 'Francisco Canaro',artist_music, music_features )
#print('artist music features:')
#print(artist_music_feature)
```
